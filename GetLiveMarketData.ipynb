# Alpha Vantage API Documentation

```python
import csv
import requests

CSV_URL = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY_EXTENDED&symbol=IBM&interval=30min&slice=year2month12&apikey=VS868RQM0AY6XIX7'

with requests.Session() as s:
    download = s.get(CSV_URL)
    decoded_content = download.content.decode('utf-8')
    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    my_list = list(cr)
    for row in my_list:
        print(row)
```

```python
df = pd.DataFrame(my_list)
```
